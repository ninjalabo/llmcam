```python
#| hide
#| eval: false
BASE_URL = "https://rata.digitraffic.fi"

# Load and parse the OAS file
with open("openapi.json") as f:
    spec_dict = json.load(f)
# oas = OpenAPI.parse_obj(spec_dict)
oas = spec_dict
tool_schema = toolbox_schema(BASE_URL, oas)
```

```python
#| eval: false
messages = [
    form_msg("system", "You are a helpful system administrator. Use the supplied tools to help the user."),
    form_msg("user", "Hello, what can you provide me about railway and train information?")
]
complete(messages, tool_schema, aux_fn=generate_request)
for message in messages:
    if message['role'] != "tool" and message['content']:
        print(f">> {message['role'].capitalize()}:")
        print(message["content"], "\n")
```
